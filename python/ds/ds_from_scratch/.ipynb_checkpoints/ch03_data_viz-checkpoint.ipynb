# Creating line chart

```python
from matplotlib import pyplot as plt


# data
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]



### Create line chart
plt.plot(years, gdp, color="green")
plt.show()
```

```python
plt.plot(years, gdp, color="green", marker="o")
plt.show()
```

```python

plt.plot(years, gdp, color="green", marker="o", linestyle="--", markerfacecolor='red')
plt.grid(True)
plt.title("Nominal GP")
plt.ylabel("Billions of $")
plt.show()
```

## Bar Chart
bar chart is good choice when you wnat to show how some quatity varies amnog some discrete set of items<br>

award: discrete<br>
movie: variety

```python
from matplotlib import pyplot as plt

# data
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

plt.bar(range(len(movies)), num_oscars)
plt.title("My Favorite Movies")
plt.ylabel("# of academy awards")
# labling x axis
plt.xticks(range(len(movies)), movies)
plt.show()
```

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
# Bucket grades by decile, but put 100 in with the 90s
# create group 10 -> 87, 82, 83, 85 -> will goes to 80 group. 80: 4
# 90 is the max group 
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
histogram
```

```python
x_values = [x + 5 for x in histogram.keys()]
# x_values = histogram.keys()
y_values = histogram.values()

#print(x_values)
# 10 is the width of the bar : numbers in x seperated by 10
# (0, 0, 0) is the edge color

plt.bar(x_values, y_values, 10, edgecolor="#121212")

plt.xticks(range(10, 110, 10))
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```
